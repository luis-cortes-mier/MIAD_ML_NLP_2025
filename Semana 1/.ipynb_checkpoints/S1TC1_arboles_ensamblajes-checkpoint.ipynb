### Punto 3 - Regresión lineal
En la celda 3 ajuste un modelo de regresión lineal a un conjunto de train de datos, utilizando "total" como variable de respuesta y "season" y "hour" como las únicas variables predictoras, teniendo en cuenta que la variable "season" es categórica. Luego, imprima los coeficientes e interprételos. ¿Cuáles son las limitaciones de la regresión lineal en este caso?

```python
# Celda 3

# Como la variable "season" es categorica y no es una variable ordinal, procedo a convertirlas en variables dummies
bikes_with_dummies = pd.get_dummies(bikes, columns=['season'], drop_first= True)

# Selecciono las variables predictoras y la variable de respuesta
X = bikes_with_dummies[['season_2', 'season_3', 'season_4', 'hour']]
y = bikes_with_dummies["total"]

# Hago la particion de train y test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1111)

# Ajusto y entreno el modelo de regresion
lr = LinearRegression()
lr.fit(X_train, y_train)

# Hago predicciones
pred = lr.predict(X_test)

# Evaluo el modelo
mse = mean_squared_error(y_test, pred)
rmse = mse**0.5
r2 = r2_score(y_test, pred)

# Resultados
print("El MSE es igual a: " + str(mse))
print("El RMSE es igual a: " + str(rmse))
print("El R2 es igual a:" + str(r2))
print("Los coeficientes del modelo son: " + str(lr.coef_))
```

Conclusiones

Nota importante para comprender los coeficientes: se trabajo con variables dummies para la variable "season" y se utilizo el valor de "True" para el párametro de "dropfirst".

Interpretación de los coeficientes:
- En season_2 (primavera), el número de usuarios aumenta en 100.68 en comparación con la estación de referencia (season_1)
- En season_3 (verano), el número de usuarios aumenta en 118.55 en comparación con la estación de referencia (season_1)
- En season_4 (otoño), el número de usuarios aumenta en 81.21 en comparación con la estación de referencia (season_1)
- El coeficiente de hour de 10.63 nos indica que por cada hora adicional, el número de usuarios aumenta en promedio 10.63. Es decir, la hora del día tiene un impacto en la demanda.

### Punto 5 - Árbol de decisión con librería
En la celda 5 entrene un árbol de decisiones con la **librería sklearn**, usando las variables predictoras "season" y "hour" y calibre los parámetros que considere conveniente para obtener un mejor desempeño. Recuerde dividir los datos en conjuntos de entrenamiento y validación para esto. Comente el desempeño del modelo con alguna métrica de desempeño de modelos de regresión y compare desempeño con el modelo del punto 3.

```python
# Celda 5

# Voy a utilizar el dataframe "bikes" definido desde un inicio
# En este ejercicio no haré uso de las variables dummies (No es estrictamente necesario en arboles de decisiones)

# Defino mis variables predictoras y la variable de respuesta 
X = bikes[["season", "hour"]]
y = bikes["total"]

# Hago la particion de train y test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1111)

# Voy a calibrar el árbol pre poda con gridsearch. Calibraré el hiperpárametro de maxdepth

# Defino los valores posibles de maxdepth
grid_parametros = {"max_depth": [2, 4, 6, 8]}

# Creo el modelo
td = DecisionTreeRegressor(random_state = 42)

# Realizo la busqueda en el grid 
busqueda_grid = GridSearchCV(estimator = td, param_grid = grid_parametros, cv = 5, scoring = 'neg_mean_squared_error')

# Ajusto el modelo con train
busqueda_grid.fit(X_train, y_train)

# Mejor párametro
mejor_parametro = busqueda_grid.best_params_
print("Los mejores párametros encontrados son: " + str(mejor_parametro))

# Árbol con mejor parametro
mejor_modelo = busqueda_grid.best_estimator_

# Evaluo el modelo
predicciones = mejor_modelo.predict(X_test)

# Calculo MSE y R2
mse = mean_squared_error(y_test, predicciones)
r2 = r2_score(y_test, predicciones)

# Imprimo los resultados 
print("El MSE es igual a: " + str(mse))
print("El R2 es igual a:" + str(r2))

```

Con este modelod de árbol de decisión se obtuvo un mejor rendimiento que con la regresión lineal. Seguramente las relaciones entre los datos no son totalmente lineales, y eso es un punto negativo para el modelo de regresion usado en el incisio #3. A pesar de que el árbol de decision si pudo capturar esas relaciones no lineales, no tuvo un desempeño óptimo por lo que se hace necesario buscar otras alternativas como el ensamblaje de modelos.

De manera general podriamos decir que el modelo de regresión lineal solo es capaz de explicar la variación de los datos en aproximadamente 21%, mientras que el modelo basado en árbol de decisión (con una profundidad máxima de 8) puede explicar la variabilidad de los datos en un 59%.

### Datos popularidad de noticias

```python
# Lectura de la información de archivo .csv
df = pd.read_csv('https://raw.githubusercontent.com/davidzarruk/MIAD_ML_NLP_2023/main/datasets/mashable.csv', index_col=0)
df.head()
```

```python
# Definición variable de interes y variables predictoras
X = df.drop(['url', 'Popular'], axis=1)
y = df['Popular']
y.mean()
```

```python
# División de la muestra en set de entrenamiento y prueba
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

### Punto 6 - Árbol de decisión y regresión logística
En la celda 6 construya un árbol de decisión y una regresión logística. Para el árbol calibre al menos un parámetro y evalúe el desempeño de cada modelo usando las métricas de Accuracy y F1-Score.

```python
# Celda 6

# Modelo de arbol de decision 

# Calibraré el hiperpárametro de maxdepth y min_samples_leaf

grilla_parametros = {"max_depth": [2, 4, 6, 8],
                     "min_samples_leaf": [1, 2, 4, 6] }

# Creo el modelo
td1 = DecisionTreeClassifier(random_state = 42)

# Realizo la busqueda en el grid 
busqueda_grid_1 = GridSearchCV(estimator = td1, param_grid = grilla_parametros, cv = 5, scoring = 'accuracy')

# Ajusto el modelo con train
busqueda_grid_1.fit(X_train, y_train)

# Mejores párametros
mejores_parametros_td1 = busqueda_grid_1.best_params_

# Árbol con mejores parametros
mejor_modelo_td1 = busqueda_grid_1.best_estimator_

# Predigo con el mejor modelo y sus mejores párametros
predicciones_td1 = mejor_modelo_td1.predict(X_test)

# Evaluo el modelo
accuracy_td1 = accuracy_score(y_test, predicciones_td1)
f1_td1 = f1_score(y_test, predicciones_td1, average = "weighted")

# ________________________________________________________________________________

# Modelo de regresión lógistica

# creo el modelo
lr = LogisticRegression(random_state=42)

# Ajusto el modelo
lr.fit(X_train, y_train)

# Predigo con el modelo
predicciones_rl = lr.predict(X_test)

# Evaluo el modelo con accuracy y f1-score
accuracy_rl = accuracy_score(y_test, predicciones_rl)
f1_rl = f1_score(y_test, predicciones_rl, average = "weighted")

# ________________________________________________________________________________

# Resultados finales
print("ARBOL DE DECISIÓN\n") 
print("Los mejores párametros encontrados son: " + str(mejores_parametros_td1))
print("Accuracy = " + str(accuracy_td1))
print("F1 Score = " + str(f1_td1))

print("\n")

print("REGRESIÓN LÓGISTICA\n") 
print("Accuracy = " + str(accuracy_rl))
print("F1 Score = " + str(f1_rl))
```

Conclusiones:

El árbol de decisión construido tuvo un mejor desempeño tanto con el accuracy como con el F1 Score que el modelo de regresión lógistica. Aqui es importante denotar la importancia de calibrar un modelo con sus hiperpárametros. En este ejercicio, la calibración se realizo pre-poda, es decir, se hizo durante la construcción del modelo. La precisión del modelo no es la mejor, y seguramente puede ser mejorado.

### Punto 7 - Votación Mayoritaria
En la celda 7 elabore un esamble con la metodología de **Votación mayoritaria** compuesto por 300 muestras bagged donde:

-las primeras 100 muestras vienen de árboles de decisión donde max_depth tome un valor de su elección\
-las segundas 100 muestras vienen de árboles de decisión donde min_samples_leaf tome un valor de su elección\
-las últimas 100 muestras vienen de regresiones logísticas

Evalúe cada uno de los tres modelos de manera independiente utilizando las métricas de Accuracy y F1-Score, luego evalúe el ensamble de modelos y compare los resultados. 

Nota: 

Para este ensamble de 300 modelos, deben hacer votación mayoritaria. Esto lo pueden hacer de distintas maneras. La más "fácil" es haciendo la votación "manualmente", como se hace a partir del minuto 5:45 del video de Ejemplo práctico de emsablajes en Coursera. Digo que es la más fácil porque si hacen la votación mayoritaria sobre las 300 predicciones van a obtener lo que se espera.

Otra opción es: para cada uno de los 3 tipos de modelos, entrenar un ensamble de 100 modelos cada uno. Predecir para cada uno de esos tres ensambles y luego predecir como un ensamble de los 3 ensambles. La cuestión es que la votación mayoritaria al usar los 3 ensambles no necesariamente va a generar el mismo resultado que si hacen la votación mayoritaria directamente sobre los 300 modelos. Entonces, para los que quieran hacer esto, deben hacer ese último cálculo con cuidado.

Para los que quieran hacerlo como ensamble de ensambles, digo que se debe hacer el ensamble final con cuidado por lo siguiente. Supongamos que:

* para los 100 árboles del primer tipo, la votación mayoritaria es: 55% de los modelos predicen que la clase de una observación es "1"
* para los 100 árboles del segundo tipo, la votación mayoritaria es: 55% de los modelos predicen que la clase de una observación es "1"
* para las 100 regresiones logísticas, la votación mayoritaria es: 10% de los modelos predicen que la clase de una observación es "1"

Si se hace la votación mayoritaria de los 300 modelos, la predicción de esa observación debería ser: (100*55%+100*55%+100*10%)/300 = 40% de los modelos votan porque la predicción debería ser "1". Es decir, la predicción del ensamble es "0" (dado que menos del 50% de modelos predijo un 1).

Sin embargo, si miramos cada ensamble por separado, el primer ensamble predice "1", el segundo ensamble predice "1" y el último ensamble predice "0". Si hago votación mayoritaria sobre esto, la predicción va a ser "1", lo cual es distinto a si se hace la votación mayoritaria sobre los 300 modelos.

```python
# Celda 7

## Defino los párametros a utilizar y creo los 3 modelos con su respectivo párametro

max_depth = 5
min_samples_leaf = 5

# 1er modelo: árbol de decisión calibrado con max_depth
dt_max_depth = DecisionTreeClassifier(max_depth = max_depth, random_state = 42)

# 2do modelo: árbol de decisión calibrado con min_samples_leaf
dt_min_samples_leaf = DecisionTreeClassifier(min_samples_leaf = min_samples_leaf, random_state = 42)

# 3er moodelo: Regresión Lógistica
log_reg = LogisticRegression(random_state = 42)


## Genero las 100 muestras bagged para cada modelo

# Muestras bagged para 1er modelo
X_dt_max_depth, Y_dt_max_depth = resample(X_train, y_train, n_samples=100 ,random_state=42)

# Muestras bagged para 2do modelo
X_dt_min_samples_leaf, Y_dt_min_samples_leaf = resample(X_train, y_train, n_samples=100, random_state=42)

# Muestras bagged para 3er modelo
X_log_reg, Y_log_reg = resample(X_train, y_train, n_samples=100, random_state=42)


## Entrenamiento de modelos con las muestras generadas anteriormente 

dt_max_depth.fit(X_dt_max_depth, Y_dt_max_depth)
dt_min_samples_leaf.fit(X_dt_min_samples_leaf, Y_dt_min_samples_leaf)
log_reg.fit(X_log_reg, Y_log_reg)


## Realizo las predicciones con cada modelo

pred_dt_max_depth = dt_max_depth.predict(X_test)
pred_dt_min_samples_leaf = dt_min_samples_leaf.predict(X_test)
pred_log_reg = log_reg.predict(X_test)


## Evaluo los modelos

print("Árbol con max_depth\n")
print("Accuracy = " + str(accuracy_score(y_test, pred_dt_max_depth)))
print("F1 Score = " + str(f1_score(y_test, pred_dt_max_depth, average = "weighted")))

print("\n")

print("Árbol con min_samples_leaf\n")
print("Accuracy = " + str(accuracy_score(y_test, pred_dt_min_samples_leaf)))
print("F1 Score = " + str(f1_score(y_test, pred_dt_min_samples_leaf, average = "weighted")))

print("\n")

print("ÁRegresión Lógistica\n")
print("Accuracy = " + str(accuracy_score(y_test, pred_log_reg)))
print("F1 Score = " + str(f1_score(y_test, pred_log_reg, average = "weighted")))

# _________________________________________________________________________________________________

## Ensamblo los modelos con votación mayoritaria 

vot_may = VotingClassifier(estimators = [("dt_max_depth",dt_max_depth), ("dt_min_samples_leaf", dt_min_samples_leaf), ("log_reg", log_reg)], voting= "hard")

# Entreno el ensamblaje
vot_may.fit(X_train, y_train)

# Predigo con el ensamblaje entrenado
pred_ensam = vot_may.predict(X_test)

# Evaluo el modelo de ensamblaje

print("\n")

print("Modelo de ensamblaje por votación mayoritaria \n")
print("Accuracy = " + str(accuracy_score(y_test, pred_ensam)))
print("F1 Scores = " + str(f1_score(y_test, pred_ensam)))
```

### Punto 8 - Votación Ponderada
En la celda 8 elabore un ensamble con la metodología de **Votación ponderada** compuesto por 300 muestras bagged para los mismos tres escenarios del punto 7. Evalúe los modelos utilizando las métricas de Accuracy y F1-Score

```python
# Celda 8

# En este punto voy a utilizar los modelos entrenados y ajustados en el punto 7

# Nota: para la asignación de pesos, utilizaré el accuracy obtenido en el punto 7
accuracy_mod_1 = accuracy_score(y_test, pred_dt_max_depth)
accuracy_mod_2 = accuracy_score(y_test, pred_dt_min_samples_leaf)
accuracy_mod_3 = accuracy_score(y_test, pred_log_reg)


# Asigno pesos a cada modelo
pesos = {"dt_max_depth":accuracy_mod_1, "dt_min_samples_leaf":accuracy_mod_2, "log_reg":accuracy_mod_3}

## Creo el ensamble con votación ponderada

# Array para almacenar predicciones ponderadas
predicciones_ponderadas = np.zeros_like(y_test, dtype=float)

# Cantidad de modelos
num_modelos = len(pesos)

# Genero un ciclo for para hacer las operaciones de ponderación en cada modelo
for modelo, peso in pesos.items():
    if modelo == "dt_max_depth":
        predicciones_ponderadas += peso * (pred_dt_max_depth == 1)
    elif modelo == "dt_min_samples_leaf":
        predicciones_ponderadas += peso * (pred_dt_min_samples_leaf == 1)
    elif modelo == "log_reg":
        predicciones_ponderadas += peso * (pred_log_reg == 1)

predicciones_finales = (predicciones_ponderadas > 0).astype(int)

# Evaluo el ensamblaje
f1_ponderado = f1_score(y_test, predicciones_finales, average='weighted')
accuracy_ponderado = accuracy_score(y_test, predicciones_finales)

# Resultados
print("Modelo de ensamblaje por votación ponderada \n")
print("Accuracy = " + str(accuracy_ponderado))
print("F1 Scores = " + str(f1_ponderado))


```

### Punto 9 - Comparación y análisis de resultados
En la celda 9 comente sobre los resultados obtenidos con las metodologías usadas en los puntos 7 y 8, compare los resultados y enuncie posibles ventajas o desventajas de cada una de ellas.

Celda 9

Los resultados del accuracy y f1-score son mejores en el ensamble por votación mayoritaria que en el ensamble por votación ponderada. Quizás pueda parecer un poco contraintuitivo que estos resultados se presenten si nos ceñimos a la lógica que utiliza cada tipo de ensamble para realizar predicciones. Estos resultados pueden obedecer a que en el ensamblaje por votación mayoritaria se permite que los modelos con bajo rendimiento (bajo accuracy) tengan la misma influencia que los modelos con un buen rendimiento. 

Las ventajas que nos brinda la votación mayoritaria son representadas en ensamblajes más robustos en donde los modelos con peor rendimiento se “pierdan” siempre y cuando la mayoría de modelos predigan correctamente una clase. Otra de las ventajas que nos representa, puede verse reflejada en una mayor simplicidad del modelo, ya que no requiere una asignación de pesos específicos. Sin embargo, la forma en como funciona la votación mayoritaria trae consigo ciertas desventajas; una de ellas es que si existen modelos con grandes diferencias de rendimiento, los peores tendrán un efecto igual de importante que aquellos modelos con buen rendimiento.

Por otro lado, en el ensamblaje por votación ponderada los modelos tienen un peso basado en sus rendimientos (accuracy en este caso). En este sentido, los modelos con mejor rendimiento tienen un efecto mayor en la predicción. Una de las ventajas de estos modelos es que pueden generar rendimientos mayores al dar mas peso a los mejores modelos. 

En conclusión, los 3 modelos con los que se construyeron los ensamblajes tienen un rendimiento considerablemente bajo, lo que afecta a la votación ponderada y da una ligera ventaja a la votación mayoritaria.
